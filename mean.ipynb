How to find mean by groupby()

```python
import pandas as pd
```

```python
df = pd.read_csv(r'/Users/bhavishyasrivastava/Desktop/data_analyst/world_population.csv')
df
```

```python
df.columns
```

```python
# population_columns = [
#     '2022 Population', '2020 Population', '2015 Population',
#     '2010 Population', '2000 Population', '1990 Population',
#     '1980 Population', '1970 Population'
# ]

cm = df.groupby('Continent')[df.columns[5:13]].mean().sort_values(by="2022 Population", ascending=False)
cm
```

```python
df2 = cm.plot()
df2
```

```python
df2 = cm.transpose()
df2
```

```python
df2.plot()
```
