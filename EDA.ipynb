Exploratory Data Analysis in Pandas 

```python
import pandas as pd
import seaborn as slt
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r"/Users/bhavishyasrivastava/Desktop/data_analyst/world_population.csv")
df
```

```python
#s used in pandas to customize the way floating-point numbers are displayed in the output
pd.set_option('display.float_format', lambda x:'%.2f' %x)
```

```python
#to get information about the dataframes
df.info()
```

```python
df.describe()
```

```python
#used to identify null value in every columns
df.isnull().sum()
```

```python
#count unique values
df.nunique()
```

```python
#sorting on values
#head() is used to show 1st few values of dataframes
df.sort_values(by='2022 Population').head(10)
```

```python
numerical_df = df.select_dtypes(include=[np.number])
correlation_matrix = numerical_df.corr()
correlation_matrix
```

```python
#heatmap using seaborn
slt.heatmap(numerical_df.corr(), annot=True)

#to increase size using matplotlib.pyplot
plt.rcParams['figure.figsize'] = (16, 8)
plt.show()
```
