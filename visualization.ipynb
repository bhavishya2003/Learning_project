```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#to change style

print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv(r"/Users/bhavishyasrivastava/Desktop/data_analyst/Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
#for line plot diagram
df.plot(kind = 'line')
```

```python
#seggregated line plot
df.plot(kind = 'line', subplots = True)
```

```python
#adding title and 
df.plot(kind = 'line',title = 'Ice Cream Rating', xlabel = 'Daily Rating', ylabel = 'Scones')
```

```python
#plot bar diagram
df.plot(kind = 'bar')
```

```python
#stacked bar diagram
df.plot(kind = 'bar', stacked=True)
```

```python
# specific plot diagram
df['Flavor Rating'].plot(kind = 'bar')
```

```python
#horizontal bar diagram
df.plot.barh(stacked = True)
```

```python
#scatter diagram
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=300, c='Pink')
```

```python
#histagram diagram
df.plot.hist(bins=20)
```

```python
#Box diagram
df.boxplot()
```

```python
#plot area diagram
df.plot.area()
df.plot.area(figsize = (5,10))
```

```python
#pie chat
df.plot.pie(y='Flavor Rating', figsize=(10,30))
```
