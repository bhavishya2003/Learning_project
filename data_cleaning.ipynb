```python
import pandas as pd
```

```python
df = pd.read_excel(r'/Users/bhavishyasrivastava/Desktop/data_analyst/Customer Call List.xlsx')
df
```

```python
# #to drop duplicate value
df = df.drop_duplicates()
# df
```

```python
#to drop particular column
df = df.drop(columns='Not_Useful_Column')
# df
```

```python
#strip irrelevent characters
# df['Last_Name']=df['Last_Name'].str.lstrip('/')
# df['Last_Name']=df['Last_Name'].str.lstrip('...')
# df['Last_Name']=df['Last_Name'].str.rstrip('_')
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
# df
```

```python
#to replace certain character which other
#regex=True: This is required starting from Pandas version 1.3.0 to explicitly indicate that a regex(short for Regular Expression) is being used.
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
# df
```

```python
#used to add charecter in string
#The lambda keyword in Python is used to create anonymous (unnamed) functions. These functions are small and defined in a single line
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# df
```

```python
#to replace the data with another data
df["Phone_Number"] =df['Phone_Number'].str.replace('nan--','', regex=True)
df["Phone_Number"] =df['Phone_Number'].str.replace('Na--','', regex=True)
# df
```

```python
#to split the columns
df[["Street", "States", "Zip_code"]] = df["Address"].str.split(',', n=2, expand = True)
# df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y', regex=True)
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N', regex=True)
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y', regex=True)
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N', regex=True)
# df
```

```python
#replace NaN/ Na to nothing
df = df.fillna('')
# df
```

```python
#use to drop multiple row which satisfy the condition
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == "Y":
        df.drop(x, inplace = True)
df
```

```python
#use to drop multiple row which satisfy the condition
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)
```
